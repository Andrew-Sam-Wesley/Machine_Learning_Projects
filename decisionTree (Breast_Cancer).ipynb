```python
#train test split 
```

```python
#independent and dependent data
X = final_data.iloc[:,:-1].values
Y = final_data.iloc[:,-1].values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=14)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini',random_state=14,ccp_alpha=0.0)
```

```python
dt.fit(X_train,Y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
#confusion matrix
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
dt.score(X_train,Y_train)
```

```python
dt.score(X_test,Y_test)
```

```python
#pruning
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
tree.plot_tree(dt)
```

```python
data = dt.cost_complexity_pruning_path(X_train,Y_train)
ccp_alphas,impurities=data.ccp_alphas,data.impurities
```

```python
ccp_alphas
```

```python
treee = []
for i in ccp_alphas:
    dt = DecisionTreeClassifier(random_state=10,ccp_alpha=i)
    dt.fit(X_train,Y_train)
    treee.append(dt)
print("number of nodes in the last tree",treee[-1].tree_.node_count)
print(" the value of ccpalpha is ",ccp_alphas[-1])
```

```python
fig,ax=plt.subplots()

accuracy = []
acc_test = []

for i in treee:
    accuracy.append(i.score(X_train,Y_train))

for i in treee:
    acc_test.append(i.score(X_test,Y_test))
plt.grid()
ax.plot(ccp_alphas[:-1], accuracy[:-1], marker="o", drawstyle="steps-post")
ax.plot(ccp_alphas[:-1], acc_test[:-1], marker="o", drawstyle="steps-post")
```

```python
#final model with ccpalpha value of 0.011
```

```python
dt_final = DecisionTreeClassifier(random_state=10,ccp_alpha=0.011)
```

```python
dt_final.fit(X_train,Y_train)
```

```python
y_p = dt_final.predict(X_test)
```

```python
print(classification_report(Y_test,y_p))
```
